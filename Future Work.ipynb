# Qué sucedería si cada nodo observa repetidas veces a sus vecinos? 

```python
import numpy as np
import numpy.linalg as linalg
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.ladder_graph(7)
```

```python
A = nx.convert_matrix.to_numpy_matrix(G)
```

```python
vals, vec = linalg.eigh(A)
A
```

```python
row = 0
print(np.abs(vec[row]))
colors = ['blue' if (np.abs(vec.item(row,i)) > 0.3) else 'red' for i in range(0,A.shape[0])]
print(colors)
```

```python
nx.draw(G, with_labels=True, font_weight='bold', node_color=colors)
```
